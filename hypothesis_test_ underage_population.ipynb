```python
def campare(s1, s2):
    for s in s1:
        if s not in s2:
            return False
    return True
```

```python
visited = [0] * len(underage_population_2021)
for i in sorted_dong_names:
    for j in range(len(underage_population_2021)):
        
        if visited[j]:
            continue
        
        if campare(i, underage_population_2021.iloc[j, 0]):
            underage_population_2021.iloc[j, 0] = i
            visited[j] = 1
        
        
```
